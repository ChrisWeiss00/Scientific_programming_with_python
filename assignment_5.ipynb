Task 1:

Creating two user-defined function, that encode the Root-Mean-Squared-Error(RMSE). <br>
One of this function's is using the regular Python lists. <br>
The other one is using Numpy and it's functions.  <br>
Both user defined functions are computing the same result. The only difference is the runtime, which is calculated in the Task Nr. 3.

```python
import numpy as np 
import math

def RMSE_without_np(calculated,experiment):
    mean_square_error= 0
    index= 0
    for x in calculated:
        mean_square_error= mean_square_error+ (x-experiment[index])**2
    mean_square_error= mean_square_error/len(calculated)
    return math.sqrt(mean_square_error)

def RMSE_with_np(calculated, experiment):
    mean_square_error=np.square(np.subtract(calculated,experiment)).mean()
    return np.sqrt(mean_square_error)
```

Task 4:

Timing how much time the RMSE-function needs by 20,000 repetitions. <br>
The result show much more efficient it is to work directly on the Numpy-Array, instead of going the traditional way.

```python
import timeit

calculated_density=[0.995659]*len(density)
experiment_density=density['Density'].values.tolist()

print(f'The time used with the math instructions is\
    {timeit.timeit(lambda: RMSE_without_np(calculated_density,experiment_density),number=20000)} seconds')
print(f'The time used ith the numpy istructions is\
    {timeit.timeit(lambda: RMSE_with_np(calculated_density,experiment_density),number=20000)}seconds')
```
