Task 1
Calculation of the herd immunity for the diseases in the following table using the given formula:

Threshold = 1- 1/R0

|Disease|R0|Disease|R0|
|---|---|---|---|
|Common cold|3.0|Polio|7.0|
|Influenza(seasonal strains)|1.3|Chickenpox|13.|
Measles|18.|Ebola|1.8|
SARS-CoV-2(Delta strain)|5.1|Smallpox|6.0|



```python
def calculate_threshold(r):
    pc=1-(1/r)
    return round(pc*100)

r0Dictionary ={ 'Common cold' : 3.0,                  'Polio' : 7.0,
                'Influenza(seasonal strains)' : 1.3,  'Chickenpox' : 13,
                'Measles' : 18,                      'Ebola' : 1.8,
                'SARS-CoV-2(delta-strain)' : 5.1,     'Smallpox' : 6.0}

for name in r0Dictionary:
    threshold_value=calculate_threshold(r0Dictionary[name])
    print(f'The herd immunity threshold value for {name} is {threshold_value}%.')

```
