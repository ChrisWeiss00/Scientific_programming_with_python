Computing the herd immunity threshold value, as a percentage, based on different reproduction numbers



```python
herd_immunity=0

reproduction_numbers=[2.2, 5.08]
for reproduction_number in reproduction_numbers:
    herd_immunity=1-(1/reproduction_number)
    herd_immunity=round(herd_immunity, 4)
    print(f'{herd_immunity}% have to immune, if the reproduction number is {reproduction_number}, for the herd immunity.')
```
